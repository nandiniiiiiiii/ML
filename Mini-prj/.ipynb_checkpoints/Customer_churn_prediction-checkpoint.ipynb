```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from tensorflow import keras 
import tensorflow as tf
```

```python
df = pd.read_csv('Customer_churn.csv')
df.head()
```

```python
df.drop('customerID',axis='columns',inplace=True)
df.dtypes
```

```python
# removing null val
df1 = df[df.TotalCharges!=' ']
df1.shape
```

```python
df.dtypes
```

```python
df1.TotalCharges=pd.to_numeric(df1.TotalCharges)
```

```python
df1.TotalCharges.dtypes
```

```python
tenure_no = df1[df1.Churn=="No"].tenure
tenure_yes = df1[df1.Churn=="Yes"].tenure

plt.xlabel("tenure")
plt.xlabel("No. od customers")
plt.title("Customer_churn_prediction")

plt.hist([tenure_yes,tenure_no],color=['green','red'],label=['chunk_yes','chunk_no'])
plt.legend()
```

```python
for column in df1:
    if df1[column].dtype=='object':
        print(f'{column}:{df1[column].unique()}')
```

```python
df1.replace("No internet service","No",inplace=True)
df1.replace("No phone service","No",inplace=True)
```

```python
for column in df1:
    if df1[column].dtype=='object':
        print(f'{column}:{df1[column].unique()}')
```

```python
yes_no_col = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection',
              'TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
for col in yes_no_col:
    df1[col].replace({'Yes':1,'No':0},inplace=True)
```

```python
for column in df1:
    print(f'{column}:{df1[column].unique()}')
```

```python
from sklearn import preprocessing
```

```python
df1['gender'].replace({'Female':1,'Male':0},inplace=True)
```

```python
for column in df1:
    print(f'{column}:{df1[column].unique()}')
```

```python
ohc = preprocessing.OneHotEncoder()
ohc.fit(df1[['InternetService']])
ohc.categories_
```

```python
one_hot = ohc.transform(df1[['InternetService']]).toarray()
one_hot
```

```python
df1[['DSL', 'Fiber optic', 'No']] = one_hot
df1.head()
```

```python
df1.head()
```

```python
df2 = pd.get_dummies(data=df1,columns=['Contract','PaymentMethod'])
df2.head()
```

```python
df2.drop('InternetService',axis='columns',inplace=True)
```

```python
df2.dtypes
```

```python
df2[['DSL', 'Fiber optic', 'No']] = df2[['DSL', 'Fiber optic', 'No']].astype(int)
```

```python
df2.dtypes
```

```python
# changing range between o-1
col = ['MonthlyCharges','TotalCharges','tenure']
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2[col] = scaler.fit_transform(df2[col])
```

```python
df2['tenure'].unique()
```

```python
# train-test split
x = df2.drop('Churn',axis='columns')
y = df2['Churn']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
model = keras.Sequential([
    keras.layers.Dense(20, input_shape=(26,),activation='relu'),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(x_train,y_train,epochs=100)
```

```python
model.evaluate(x_test,y_test)
```

```python
yp = model.predict(x_test)
yp
```

```python
yp_pred=[]
for ele in yp:
    if ele>0.5:
        yp_pred.append(1)
    else:
        yp_pred.append(0)
```

```python
yp_pred[:5]
```

```python
y_test[:5]
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = tf.math.confusion_matrix(labels=y_test,predictions=yp_pred)
```
